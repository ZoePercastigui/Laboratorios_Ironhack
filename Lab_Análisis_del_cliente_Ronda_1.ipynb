- Concatenar los tres marcos de datos

```python
# ahora bien ya que tenemos por el momento estandarizadas y organizadas nuestras columnas 
# 'file1', 'file2' y 'file3' las vamos a concatenar usando pd.concat 
# los vamos a apilar verticalmente (es decir, se concatenaran por filas) usando 'axis=0' 
# lo que resulta en un nuevo DF el cual llamaremos data 
data = pd.concat([file1,file2,file3], axis=0)
data
```

```python
# continuando con la limpieza de nuestros datos , aqui estandarizamos los nombres de las columnas 
# y eliminamos los espacios en blanco remplazandolos con un "_"
data = data.rename(columns=lambda x: x.strip().replace(' ', '_'))
data
```

```python
# Revisando los datos que se encuentran en la columna de 'ST' 
# nos damos cuenta que no se encuentran homogeneizados
data['ST'].unique()
```

```python
#revisando la frecuencia de mis datos de la columna 'ST' 
data['ST'].value_counts()
```

```python
# para limpiar la columan 'ST', lo que me interesa es que me aparezcan los nombres completos del estado 
# para ello crearemos una función que nos permita realizar esta limpieza
def clean_ST(celda):
  if celda in ['AZ']:
    return 'Arizona'
  elif celda in ['Cali']:
    return 'California'
  elif celda in ['WA']:
    return 'Washington'
  else:
    return celda
```

```python
data['ST']=list(map(clean_ST,data['ST']))
data['ST']
```

```python
# aqui ya homogeneizamos los datos de nuestra columna 'ST' y procedemos de igual manera a verificar 
# una por una cada columna para nuestra limpieza del 'df'
data['ST'].value_counts()
```

```python
# Revisando la columna 'GENDER', nos damos cuenta que no estan homogeneizados, 
# por lo cual realizaremos nuevamente el paso anterior que utilizamos para homogeneizar
# la columna de 'ST' pero ahora para nuestra columna de 'GENDER'
data['GENDER'].unique()
```

```python
data['GENDER'].value_counts()
```

```python
def clean_GENDER(celda):
  if celda in ['Femal']:
    return 'F'
  elif celda in ['Male']:
    return 'M'
  elif celda in ['female']:
    return 'F'
  else:
    return celda
```

```python
data['GENDER']=list(map(clean_GENDER,data['GENDER']))
data['GENDER']
```

```python
data['GENDER'].value_counts()
```

```python
# Revisando la columna 'EDUCATION' , tenemos el mismo caso asi que aplicaremos 
# nuevamente la limpieza de la misma
data['EDUCATION'].unique()
```

```python
data['EDUCATION'].value_counts()
```

```python
def clean_EDUCATION(celda):
  if celda in ['Bachelors']:
    return 'Bachelor'
  else:
    return celda
```

```python
data['EDUCATION']=list(map(clean_EDUCATION,data['EDUCATION']))
data['EDUCATION']
```

```python
data['EDUCATION'].value_counts()
```

```python
# Al revisar la columna de 'NUMBER_OF_OPEN_COMPLAINTS' nos percatamos de la misma problematica
data['NUMBER_OF_OPEN_COMPLAINTS'].unique()
```

```python
 # Revisando la frecuencia y viendo que encontramos una similitud de datos que van de 1 a 5 
 # homogeneizaremos de esta manera 
data['NUMBER_OF_OPEN_COMPLAINTS'].value_counts()
```

```python
def clean_NUMBER_OF_OPEN_COMPLAINTS(celda):
  if celda in ['1/0/00']:
    return '1'
  elif celda in ['1/1/00']:
    return '1'
  elif celda in ['1/2/00']:
    return '2'
  elif celda in ['1/3/00']:
    return '3'
  elif celda in ['1/4/00']:
    return '4'
  elif celda in ['1/5/00']:
    return '5'      
  else:
    return celda
```

```python
data['NUMBER_OF_OPEN_COMPLAINTS']=list(map(clean_NUMBER_OF_OPEN_COMPLAINTS,data['NUMBER_OF_OPEN_COMPLAINTS']))
data['NUMBER_OF_OPEN_COMPLAINTS']
```

  - Elimine la columna educación y el número de quejas abiertas del marco de datos.


```python
# Eliminamos las columnas 'EDUCATION' y 'NUMBER_OF_OPEN_COMPLAINTS'
data = data.drop(['EDUCATION','NUMBER_OF_OPEN_COMPLAINTS'], axis=1)
data
```

  - Corrija los valores en la columna valor de por vida del cliente. Se dan como un porcentaje, así que multiplíquelos por 100 y cambie `dtype` a tipo `numérico`.


```python
# para corregir los valores en la columna 'CUSTOMER_LIFETIME_VALUE' que se dan como un porcentaje, 
# multiplícarlos por 100 y posteriormente cambiar el dtype a tipo numérico.
# primero definimos una función que nos realice esta operación , 
# dentro de nuestra función indicaremos que también convierta cualquier valor no numérico en NaN.

def conv_porc(valor):
    valor = valor * 100
    valor = pd.to_numeric(valor, errors='coerce')
    return valor
```

```python
data['CUSTOMER_LIFETIME_VALUE'] = data['CUSTOMER_LIFETIME_VALUE'].apply(conv_porc)
data
```


  - Compruebe si hay filas duplicadas en los datos y elimínelas, si las hay.
 

```python
# creamos una nueva variable que se llamada 'data_duplicados' 
# la cual contiene las filas que se encuentran duplicadas de nuestro DF
data_duplicados = data.duplicated()
data_duplicados
```

```python
# Haciendo revisión de las filas duplicadas vemos que existen 2939 filas duplicadas del DF. ()
data_duplicados.value_counts()
```

```python
# Eliminando las filas duplicadas del DF 
# dando la siguiente indicación 'keep='first' le estamos indicando que la primera ocurrencia de cada fila duplicada 
# se mantenga y las siguientes se eliminarán.
# lo cual comprobamos al verificar que efectivamente en nuestra 'data' tenemos 9135 filas con 9 'columnas'
data = data.drop_duplicates(keep='first')
data
```
