```python
# Tratamos los valores nulos y los convertimos a 'not specified' de todas las columas de tipo 'object'
# los llenamos con un string 'not specified'
data[['STATE','RESPONSE','VEHICLE_CLASS','VEHICLE_SIZE','VEHICLE_TYPE']] = data[['STATE', 'RESPONSE','VEHICLE_CLASS','VEHICLE_SIZE','VEHICLE_TYPE']].fillna('not specified')
```

```python
# tratando los valores nulos de laas columnas de tipo 'object' las rellenamos con (0)
data[['MONTHS_SINCE_LAST_CLAIM','NUMBER_OF_OPEN_COMPLAINTS']] = data[['MONTHS_SINCE_LAST_CLAIM','NUMBER_OF_OPEN_COMPLAINTS']].fillna(0)
```

```python
# verificando que ya no tengamos valores nulos en la df
data.isna().sum()
```

```python
# Continuando con la limpieza del df , nos percatamos que la columna 'UNNAMED:_0' no tienen ninguna
# relevancia dejarala pues son los mismos datos que nuestro index por lo cual procederemos a borrarla
data = data.drop(['UNNAMED:_0'], axis=1)
data
```

6. Formato de fecha y hora: extraiga los meses del conjunto de datos y guárdelos en una columna separada. Luego filtre los datos para mostrar solo la información del primer trimestre, es decir. enero, febrero y marzo. _Pista_: Si no existen datos de marzo, considere solo enero y febrero.

```python
# Convertimos la columna 'EFFECTIVE_TO_DATE' en formato datetime primero ya que esta es de type 'object'
data['EFFECTIVE_TO_DATE'] = pd.to_datetime(data['EFFECTIVE_TO_DATE'])
data['MONTHS'] = data['EFFECTIVE_TO_DATE'].dt.month
data
```

```python
data.shape
```

```python
data['MONTHS'].value_counts  
# de acuerdo a lo que nos  indica python al ejecutar esta acción (IndexOpsMixin.value_counts of 0     2)
# no hay valores de marzo
```

```python
# como no existe marzo solo realizamos el filtrado de los 2 primeros meses y llamamos esta variable 'primer_trimestre'
primer_trimestre = data[(data['MONTHS'] >= 1) & (data['MONTHS'] <=2 )]
print(primer_trimestre)
```
